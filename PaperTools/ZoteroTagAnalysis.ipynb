```python
import csv
```

```python
spamreader = csv.DictReader(open('tags.csv'))
```

```python
paperList = []
tagDict = {}
countTags = 0
for row in spamreader:
    row['tags'] = row['tags'].split('; ')
    for tag in row['tags']:
        tagDict[tag.lower()] = tagDict.get(tag.lower(), 0) + 1
        countTags += 1
    paperList.append(dict(row))
```

```python
tagList = []
for key, value in tagDict.items():
    tagList.append([key, value])
```

```python
def takeCount(tag):
    return tag[1]

# 打开一个文件
fo = open("tagsCount", "w") 

tagList.sort(key=takeCount, reverse=True)
for tag in tagList:
    # print(tag[0], tag[1])
    strLine = tag[0] + " " + str(tag[1]) + "\n"
    fo.write(strLine)

# 关闭打开的文件
fo.close()
```

```python
for paper in paperList:
    score = 0
    for tag in paper.get('tags', []):
        score += tagDict.get(tag, 0)
    paper['score'] = score
    
def takeScore(item):
    return item.get('score', 0)

paperList.sort(key=takeScore, reverse=True)
for paper in paperList:
    print(paper['title'], paper['score'])
```
