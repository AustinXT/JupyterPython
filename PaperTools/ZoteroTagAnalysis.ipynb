```python
import csv
```

```python
spamreader = csv.DictReader(open('tags.csv'))
```

```python
paperList = []
tagDict = {}
countTags = 0
for row in spamreader:
    row['tags'] = row['tags'].split('; ')
    for tag in row['tags']:
        tagDict[tag] = tagDict.get(tag, 0) + 1
        countTags += 1
    paperList.append(dict(row))
# print(tagDict, countTags)
```

```python
for paper in paperList:
    score = 0
    for tag in paper.get('tags', []):
        score += tagDict.get(tag, 0)
    paper['score'] = score
    
def takeScore(item):
    return item.get('score', 0)

paperList.sort(key=takeScore)
print(paperList[0:10])
```
